```python
import torch
import warnings
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

torch.manual_seed(0)
warnings.filterwarnings("ignore", message="Port 8787 is already in use")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

```python
# ------------------------------
# Set up training & testing data
# ------------------------------

X = xr.open_dataset('/glade/work/skygale/_projects/SeaIceDownscaling/data/X_hist_buffer.nc').X.values
Y = xr.open_dataset('/glade/work/skygale/_projects/SeaIceDownscaling/data/Y_hist_buffer.nc').Y.values

# Ensure same sizes
X = np.swapaxes(X, 1, 2)

# Clip to max 3 m thickness
X = np.clip(X, None, 6.0)
Y = np.clip(Y, None, 6.0)

# Get land-sea mask
mask = (np.std(Y, axis=(0, 1, 2)) > 0).astype(float)
mask = torch.tensor(mask).float().unsqueeze(0).unsqueeze(0)

X.shape, Y.shape, mask.shape
```

```python
# ------------------------------
# Patch Extraction
# ------------------------------

def extract_patches(X, Y, low_patch_size=16, stride=8):

    patches_X = []
    patches_Y = []

    N, C, H, W = X.shape
    _, _, Hh, Wh = Y.shape

    if H < low_patch_size or W < low_patch_size:
        raise ValueError(f"Patch size {low_patch_size} too large for input ({H}, {W})")

    scale_y = Hh // H
    scale_x = Wh // W

    for n in range(N):
        for i in range(0, H-low_patch_size+1, stride):
            for j in range(0, W-low_patch_size+1, stride):

                x_patch = X[n, :, i:i+low_patch_size, j:j+low_patch_size]

                scale_i = i * scale_y
                scale_j = j * scale_x

                x_hi = low_patch_size * scale_y
                y_hi = low_patch_size * scale_x

                y_patch = Y[n, :, scale_i:scale_i+x_hi, scale_j:scale_j+y_hi]

                patches_X.append(x_patch)
                patches_Y.append(y_patch)

    return torch.stack(patches_X), torch.stack(patches_Y)


def extract_context_patches(X, Y, context_size=(21, 32), target_size=(11, 16), stride=4):

    N, C, H, W = X.shape
    _, C_out, H_hi, W_hi = Y.shape

    context_h, context_w = context_size
    target_h, target_w = target_size

    # Compute scaling factors
    scale_y = H_hi // H
    scale_x = W_hi // W

    # Padding between context and target
    pad_h = (context_h - target_h) // 2
    pad_w = (context_w - target_w) // 2

    X_patches = []
    Y_patches = []

    for n in range(N):
        for i in range(0, H - context_h + 1, stride):
            for j in range(0, W - context_w + 1, stride):

                # --- Input patch (context) ---
                x_patch = X[n, :, i:i+context_h, j:j+context_w]

                # Map to high-res indices
                scale_i = i * scale_y
                scale_j = j * scale_x

                # --- Target patch (center only) ---
                y_patch = Y[
                    n, :,
                    scale_i + pad_h * scale_y: scale_i + (pad_h + target_h) * scale_y,
                    scale_j + pad_w * scale_x: scale_j + (pad_w + target_w) * scale_x
                ]

                X_patches.append(x_patch)
                Y_patches.append(y_patch)

    X_patches = torch.stack(X_patches)
    Y_patches = torch.stack(Y_patches)

    return X_patches, Y_patches


# ------------------------------
# Train/Test Split
# ------------------------------

N, T, C, H, W = X.shape

X_fields = torch.tensor(X.reshape(N * T, C, H, W)).float()
Y_fields = torch.tensor(Y.reshape(N * T, 1, Y.shape[3], Y.shape[4])).float()

# Downscale land-sea mask to low-res grid
mask_low = F.interpolate(
    mask,
    size=(H, W),
    mode='nearest'
)

mask_low = mask_low.repeat(N * T, 1, 1, 1)
X_fields = torch.cat([X_fields, mask_low], dim=1)

indices = torch.randperm(N * T)
split = int(0.7 * N * T)

train_idx = indices[:split]
test_idx = indices[split:]

X_train_fields = X_fields[train_idx]
Y_train_fields = Y_fields[train_idx]
X_test_fields = X_fields[test_idx]
Y_test_fields = Y_fields[test_idx]

# ------------------------------
# Normalize
# ------------------------------

X_mean = X_train_fields.mean(dim=(0, 2, 3), keepdim=True)
X_std = X_train_fields.std(dim=(0, 2, 3), keepdim=True)
Y_mean = Y_train_fields.mean(dim=(0, 2, 3), keepdim=True)
Y_std = Y_train_fields.std(dim=(0, 2, 3), keepdim=True)

X_train_fields = (X_train_fields - X_mean) / (X_std + 1e-6)
X_test_fields = (X_test_fields - X_mean) / (X_std + 1e-6)
Y_train_fields = (Y_train_fields - Y_mean) / (Y_std + 1e-6)
Y_test_fields = (Y_test_fields - Y_mean) / (Y_std + 1e-6)

# ------------------------------
# Extract Patches
# ------------------------------

# Normal
X_train, Y_train = extract_patches(X_train_fields, Y_train_fields)
X_test, Y_test = extract_patches(X_test_fields, Y_test_fields)

# Context
# X_train, Y_train = extract_context_patches(X_train_fields, Y_train_fields)
# X_test, Y_test = extract_context_patches(X_test_fields, Y_test_fields)

print("X_train:", X_train.shape)
print("Y_train:", Y_train.shape)
print("X_test:", X_test.shape)
print("Y_test:", Y_test.shape)
```

```python
# ------------------------------
# Drop-In UNet
# ------------------------------

class UNet(nn.Module):

    def __init__(self, in_channels, noise_channels=1, scale_factor=4):
        super().__init__()

        self.noise_channels = noise_channels
        self.scale_factor = scale_factor

        total_in = in_channels

        def conv_block(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, padding=1),
                nn.ReLU(),
                nn.Conv2d(out_c, out_c, 3, padding=1),
                nn.ReLU(),
            )

        # Encoder
        self.enc1 = conv_block(total_in, 64)
        self.pool1 = nn.MaxPool2d(2)

        self.enc2 = conv_block(64, 128)
        self.pool2 = nn.MaxPool2d(2)

        self.enc3 = conv_block(128, 256)
        self.pool3 = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = conv_block(256, 512)

        # Decoder
        self.up3 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.dec3 = conv_block(512, 256)

        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.dec2 = conv_block(256, 128)

        self.up1 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.dec1 = conv_block(128, 64)

        # Learned high-res upsampling (NO bilinear)
        self.final_up = nn.ConvTranspose2d(
            64, 32,
            kernel_size=scale_factor,
            stride=scale_factor
        )

        self.out_conv = nn.Conv2d(32, 1, 3, padding=1)

        # FiLM generators (noise → scale/shift)
        self.film1 = nn.Conv2d(noise_channels, 64 * 2, 1)
        self.film2 = nn.Conv2d(noise_channels, 128 * 2, 1)
        self.film3 = nn.Conv2d(noise_channels, 256 * 2, 1)
        self.film_b = nn.Conv2d(noise_channels, 512 * 2, 1)

    def film(self, x, z, film_layer):
        z_resized = F.interpolate(z, size=x.shape[-2:], mode='bilinear', align_corners=False)
        gamma_beta = film_layer(z_resized)
        gamma, beta = torch.chunk(gamma_beta, 2, dim=1)
        return x * (1 + torch.tanh(gamma)) + beta

    def forward(self, x, up_size, z=None):

        if z is None:
            z = torch.randn(
                x.size(0), self.noise_channels,
                x.size(2), x.size(3),
                device=x.device
            )

        # Encoder + FiLM
        e1 = self.enc1(x)
        e1 = self.film(e1, z, self.film1)

        e2 = self.enc2(self.pool1(e1))
        e2 = self.film(e2, z, self.film2)

        e3 = self.enc3(self.pool2(e2))
        e3 = self.film(e3, z, self.film3)

        # Bottleneck
        b = self.bottleneck(self.pool3(e3))
        b = self.film(b, z, self.film_b)

        # Decoder
        d3 = self.up3(b)
        d3 = torch.cat([d3, e3], dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.dec1(d1)

        # Sharp output + match final shape
        out = self.final_up(d1)
        out = self.out_conv(out)
        out = F.interpolate(out, size=up_size, mode="bilinear", align_corners=False)

        return out


# ------------------------------
# Model Setup
# ------------------------------

model = UNet(
    in_channels=X_train.shape[1],
).to(device)

optimizer = optim.Adam(model.parameters(), lr=1e-4)

print("Done!")
```

```python
# ------------------------------
# Evaluation
# ------------------------------
model.eval()
batch_size = 16
preds = []

with torch.inference_mode():
    for i in range(0, X_test.shape[0], batch_size):

        X_batch = X_test[i:i+batch_size].to(device)
        Y_batch = Y_test[i:i+batch_size]

        Y_pred_batch = model(X_batch, Y_batch.shape[-2:])
        preds.append(Y_pred_batch.cpu())

Y_pred = torch.cat(preds, dim=0)
Y_pred_phys = Y_pred * Y_std + Y_mean
Y_test_phys = Y_test * Y_std + Y_mean

# ------------------------------
# Bilinear baseline
# ------------------------------
sit_idx = 0

X_test_sit_phys = (
    X_test[:, sit_idx:sit_idx+1] * X_std[:, sit_idx:sit_idx+1]
    + X_mean[:, sit_idx:sit_idx+1]
)

base = F.interpolate(
    X_test_sit_phys,
    size=Y_test.shape[-2:],
    mode="bilinear",
    align_corners=False
)

# ------------------------------
# Metrics
# ------------------------------
mse_cnn = ((Y_pred_phys - Y_test_phys)**2).mean().item()
mse_base = ((base - Y_test_phys)**2).mean().item()


def spatial_corr(y_true, y_pred):
    y_true_flat = y_true.reshape(y_true.shape[0], -1)
    y_pred_flat = y_pred.reshape(y_pred.shape[0], -1)

    corrs = []
    valid = 0
    total = y_true.shape[0]

    for i in range(y_true.shape[0]):

        std_true = np.std(y_true_flat[i])
        std_pred = np.std(y_pred_flat[i])

        # Skip bad patches
        if std_true > 0 and std_pred > 0:
            valid += 1
            corrs.append(np.corrcoef(y_true_flat[i], y_pred_flat[i])[0, 1])

    print(f"Valid patches: {valid}/{total}")

    return np.nanmean(corrs)


corr_cnn = spatial_corr(
    Y_test_phys.cpu().numpy(),
    Y_pred_phys.cpu().numpy()
)

corr_base = spatial_corr(
    Y_test_phys.cpu().numpy(),
    base.cpu().numpy()
)

print(f"\nCNN Test MSE         : {mse_cnn:.4f}")
print(f"Bilinear baseline MSE: {mse_base:.4f}")
print(f"\nCNN Test spatial correlation         : {corr_cnn:.4f}")
print(f"Bilinear baseline spatial correlation: {corr_base:.4f}")
```

### Training Loop - Engression

```python
# ------------------------------
# Training Loop (Engression-style)
# ------------------------------

num_epochs = 20
batch_size = 16
K = 4  # ensemble size
loss_array = []


def energy_loss(preds, y_true):
    y_true = y_true.unsqueeze(1)

    # Data fit term
    term1 = torch.mean(torch.abs(preds - y_true))

    # Diversity term (pairwise differences)
    B, K = preds.shape[:2]
    preds_i = preds.unsqueeze(2)  # (B, K, 1, ...)
    preds_j = preds.unsqueeze(1)  # (B, 1, K, ...)

    pairwise = torch.abs(preds_i - preds_j)
    term2 = pairwise.mean()

    return term1 - 0.7 * term2


def gradient_loss(pred, target):
    dx_pred = pred[:, :, :, 1:] - pred[:, :, :, :-1]
    dx_true = target[:, :, :, 1:] - target[:, :, :, :-1]

    dy_pred = pred[:, :, 1:, :] - pred[:, :, :-1, :]
    dy_true = target[:, :, 1:, :] - target[:, :, :-1, :]

    return F.l1_loss(dx_pred, dx_true) + F.l1_loss(dy_pred, dy_true)


for epoch in range(num_epochs):

    model.train()
    epoch_loss = 0.0

    idx = torch.randperm(X_train.size(0))

    for i in range(0, X_train.size(0), batch_size):

        optimizer.zero_grad()

        batch_idx = idx[i:i+batch_size]
        X_batch = X_train[batch_idx].to(device, non_blocking=True)
        Y_batch = Y_train[batch_idx].to(device, non_blocking=True)

        B, _, H, W = X_batch.shape

        # ------------------------------
        # Ensemble forward passes
        # ------------------------------
        preds = []

        for k in range(K):
            z = 0.5 * torch.randn(B, 1, H, W, device=device)  # Gaussian noise
            Y_pred = model(X_batch, Y_batch.shape[-2:], z=z)
            preds.append(Y_pred)

        preds = torch.stack(preds, dim=1)  # (B, K, 1, H_hi, W_hi)

        # ------------------------------
        # Energy loss
        # ------------------------------
        loss = energy_loss(preds, Y_batch)

        # Optional stabilization
        mean_pred = preds.mean(dim=1)
        # loss += 0.01 * F.mse_loss(mean_pred, Y_batch)
        loss += 0.1 * gradient_loss(mean_pred, Y_batch)

        # ------------------------------
        # Backprop
        # ------------------------------
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item() * X_batch.size(0)

        del X_batch, Y_batch, preds, loss

    epoch_loss /= X_train.size(0)
    loss_array.append(epoch_loss)

    print(f"Epoch {epoch+1}/{num_epochs} | Loss: {epoch_loss:.6f}")

epochs = np.arange(1, len(loss_array)+1)

plt.figure(figsize=(8, 4))
plt.plot(epochs, loss_array, marker='o', linestyle='-', color='C0', linewidth=2, markersize=6, label='Training Loss')

plt.title('EngressNet Training Loss per Epoch', fontsize=12)
plt.xlabel('Epoch', fontsize=10)
plt.ylabel('Loss', fontsize=10)
plt.xticks(epochs)
plt.xlim(0.5, len(epochs)+0.5)
plt.grid(alpha=0.25)

plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```

```python
# ------------------------------
# Evaluation
# ------------------------------
model.eval()
batch_size = 16
K_eval = 4

preds_all_full = []
preds_mean_all = []
preds_std_all = []
preds_det_all = []

with torch.inference_mode():
    for i in range(0, X_test.shape[0], batch_size):

        X_batch = X_test[i:i+batch_size].to(device)
        Y_batch = Y_test[i:i+batch_size]

        B, _, H, W = X_batch.shape

        preds = []
        for k in range(K_eval):
            z = 0.5 * torch.randn(B, 1, H, W, device=device)
            pred = model(X_batch, Y_batch.shape[-2:], z=z)
            preds.append(pred)

        preds = torch.stack(preds, dim=1)

        pred_mean = preds.mean(dim=1)
        pred_std = preds.std(dim=1)

        z0 = torch.zeros(B, 1, H, W, device=device)
        pred_det = model(X_batch, Y_batch.shape[-2:], z=z0)

        preds_all_full.append(preds.cpu())
        preds_mean_all.append(pred_mean.cpu())
        preds_std_all.append(pred_std.cpu())
        preds_det_all.append(pred_det.cpu())

preds_all_full = torch.cat(preds_all_full, dim=0)
Y_pred = torch.cat(preds_mean_all, dim=0)
Y_spread = torch.cat(preds_std_all, dim=0)
Y_pred_det = torch.cat(preds_det_all, dim=0)

Y_pred_phys = Y_pred * Y_std + Y_mean
Y_spread_phys = Y_spread * Y_std
Y_pred_det_phys = Y_pred_det * Y_std + Y_mean
Y_test_phys = Y_test * Y_std + Y_mean

# Optional quick check
plt.figure(figsize=(5, 4))
plt.imshow(Y_spread_phys[0, 0], cmap='magma')
plt.title("Prediction Uncertainty (m)")
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
# ------------------------------
# RMSE + Correlation (FIXED)
# ------------------------------

# Convert ensembles to physical units
preds = preds_all_full * Y_std + Y_mean   # (N, K, 1, H, W)
truth = Y_test_phys                       # (N, 1, H, W)
mean_pred = Y_pred_phys                   # (N, 1, H, W)

N, K, _, H, W = preds.shape

# ------------------------------
# Flatten safely
# ------------------------------
preds_flat = preds.reshape(N, K, -1)
truth_flat = truth.reshape(N, -1)
mean_flat  = mean_pred.reshape(N, -1)

# ------------------------------
# RMSE (explicit per member)
# ------------------------------
rmse_members = []

for k in range(K):
    diff = preds_flat[:, k, :] - truth_flat
    rmse_k = torch.sqrt((diff**2).mean(dim=1))  # (N,)
    rmse_members.append(rmse_k)

rmse_members = torch.stack(rmse_members, dim=1)  # (N, K)

# Ensemble mean RMSE
rmse_mean = torch.sqrt(((mean_flat - truth_flat)**2).mean(dim=1))  # (N,)

# ------------------------------
# Correlation
# ------------------------------
def corr(x, y):
    x = x - x.mean(dim=1, keepdim=True)
    y = y - y.mean(dim=1, keepdim=True)
    return (x * y).mean(dim=1) / (x.std(dim=1) * y.std(dim=1) + 1e-6)

corr_members = []

for k in range(K):
    c = corr(preds_flat[:, k, :], truth_flat)
    corr_members.append(c)

corr_members = torch.stack(corr_members, dim=1)  # (N, K)

# Ensemble mean correlation
corr_mean = corr(mean_flat, truth_flat)  # (N,)

# ------------------------------
# Aggregate statistics
# ------------------------------
rmse_members_mean = rmse_members.mean(dim=0)  # (K,)
rmse_mean_avg = rmse_mean.mean()

corr_members_mean = corr_members.mean(dim=0)  # (K,)
corr_mean_avg = corr_mean.mean()

# ------------------------------
# Print results
# ------------------------------
print("\n=== RMSE ===")
for k in range(K):
    print(f"Ensemble {k+1}: {rmse_members_mean[k]:.4f}")
print(f"Ensemble Mean: {rmse_mean_avg:.4f}")

print("\n=== Correlation ===")
for k in range(K):
    print(f"Ensemble {k+1}: {corr_members_mean[k]:.4f}")
print(f"Ensemble Mean: {corr_mean_avg:.4f}")

# ------------------------------
# 🔍 SANITY CHECKS
# ------------------------------

print("\n=== Sanity Checks ===")

# Check if ensemble members are actually different
for k in range(K - 1):
    diff = torch.abs(preds[:, k] - preds[:, k+1]).mean().item()
    print(f"Mean abs diff (Ens {k+1} vs {k+2}): {diff:.6f}")

# Check variance per member
for k in range(K):
    std_val = preds[:, k].std().item()
    print(f"Std of Ens {k+1}: {std_val:.4f}")

# Compare spread vs RMSE (calibration hint)
print("\nSpread (avg std):", Y_spread_phys.mean().item())
print("RMSE (ensemble mean):", rmse_mean_avg.item())
```
