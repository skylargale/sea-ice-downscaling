```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.patches import ConnectionPatch

kivalina_lon, kivalina_lat = -164.54, 67.73
cambridge_lon, cambridge_lat = -105.05, 69.12
```

```python
# -----------------------------------------------------------------------------
# Figure setup
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(8, 8), dpi=300)

# Main Arctic polar map
ax = plt.axes(projection=ccrs.NorthPolarStereo())
ax.set_extent([-180, 180, 55, 90], crs=ccrs.PlateCarree())

# Add features
ax.add_feature(cfeature.LAND, facecolor='grey')
ax.add_feature(cfeature.OCEAN, facecolor='white')
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linewidth=0.8)
ax.gridlines(draw_labels=False, linewidth=0.5, linestyle='--')

# -----------------------------------------------------------------------------
# Inset map: Kivalina
# -----------------------------------------------------------------------------
ax_kiv = fig.add_axes([0.52, 0.55, 0.3, 0.3], projection=ccrs.PlateCarree())

ax_kiv.set_extent([-168, -160, 65, 70], crs=ccrs.PlateCarree())
ax_kiv.add_feature(cfeature.LAND, facecolor='lightgray')
ax_kiv.add_feature(cfeature.OCEAN, facecolor='white')
ax_kiv.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax_kiv.add_feature(cfeature.BORDERS, linewidth=0.4)
ax_kiv.gridlines(draw_labels=True, linewidth=0.3, linestyle='--')

ax_kiv.plot(kivalina_lon, kivalina_lat,
            marker='o', color='red', markersize=8,
            transform=ccrs.PlateCarree())
# ax_kiv.set_title('Kivalina, Alaska')

# -----------------------------------------------------------------------------
# Inset map: Cambridge Bay
# -----------------------------------------------------------------------------
ax_cam = fig.add_axes([0.3, 0.29, 0.3, 0.3], projection=ccrs.PlateCarree())

ax_cam.set_extent([-110, -98, 67, 72], crs=ccrs.PlateCarree())
ax_cam.add_feature(cfeature.LAND, facecolor='lightgray')
ax_cam.add_feature(cfeature.OCEAN, facecolor='white')
ax_cam.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax_cam.add_feature(cfeature.BORDERS, linewidth=0.4)
ax_cam.gridlines(draw_labels=True, linewidth=0.3, linestyle='--')

ax_cam.plot(cambridge_lon, cambridge_lat,
            marker='o', color='blue', markersize=8,
            transform=ccrs.PlateCarree())
# ax_cam.set_title('Cambridge Bay, Nunavut')

# -----------------------------------------------------------------------------
# Plot locations on main map
# -----------------------------------------------------------------------------
ax.plot(kivalina_lon, kivalina_lat,
        marker='o', color='red', markersize=6,
        transform=ccrs.PlateCarree())
ax.text(kivalina_lon + 17, kivalina_lat - 1,
        'Kivalina', color='red', fontsize=10,
        transform=ccrs.PlateCarree())

ax.plot(cambridge_lon, cambridge_lat,
        marker='o', color='blue', markersize=6,
        transform=ccrs.PlateCarree())
ax.text(cambridge_lon + 6.5, cambridge_lat - 12.5,
        'Cambridge Bay', color='blue', fontsize=10,
        transform=ccrs.PlateCarree())

plt.show()
```
